# Manage null values and Encoding of categorical features

To handle null values we try to substitute them with the median for numerical features and with the mofe for categorical ones

```python
mode_val = df[categorical_feat].mode().iloc[0]
median_val = df[numerical_feat].median()
```

```python
df[categorical_feat]=df[categorical_feat].fillna(value=mode_val)
df[numerical_feat] = df[numerical_feat].fillna(value=median_val)
```

```python
df[['RainTomorrow']].isnull().sum(axis=0)
```

RainTomorrow has still null values, we remove them and we shuffle the dataset

```python
##eremove NA
df.dropna(inplace=True)
```

```python
##shuffle
df = df.sample(frac=1,random_state=1).reset_index(drop=True)
```

**Preparazione Features e della variabile Target**

Le variabili categoriche devono essere encodate




```python
##label encoder from sklearn.preprocessing for RainTdoay e RainTomorrow
le =LabelEncoder()

df[['RainToday', 'RainTomorrow']] =df[['RainToday', 'RainTomorrow']].apply(le.fit_transform)
df.head()
```

# Neural Network


Here we build the neural network, the plots_from_history function **was shared in the class**

```python
# Create two plots: one shows the accuracy values for each epoch, the other shows the loss values
# 'history' is the output of the neural network model
# 'num_epoch' is the number of epochs of the training procedure
# 'loss_name' is the name of the loss function
# 'eval_set' indicates the name of the set with which we evaluate the network performance
def plots_from_history(history, num_epoch, loss_name, eval_set):

  # Creation of the two plots
  fig, ax = plt.subplots(1,2, figsize=(18,6))
  ax = ax.ravel()

  # The first plot is about the accuracy
  ax[0].set_title("Accuracy", fontsize=16)

  # In the X-axis there are the epoch numbers
  ax[0].set_xlabel("Epoch", fontsize=14)

  # We plot in red the accuracy values related to the training set, while in blues the ones related to the test/validation set
  ax[0].plot(range(1, num_epoch + 1), history.history["accuracy"], label = "Train", c="red")
  ax[0].plot(range(1, num_epoch + 1), history.history["val_accuracy"], label = eval_set, c="blue")

  # The second plot is about the loss
  ax[1].set_title(loss_name, fontsize=16)

  # In the X-axis there are the epoch numbers
  ax[1].set_xlabel("Epoch", fontsize=14)
  
  # We plot in red the loss values related to the training set, while in blues the ones related to the test/validation set
  ax[1].plot(range(1, num_epoch + 1), history.history["loss"], label = "Train", c="red")
  ax[1].plot(range(1, num_epoch + 1), history.history["val_loss"], label = eval_set, c="blue")
  
  # Show the plots legend
  ax[1].legend(["Train", eval_set], fontsize=14, loc="upper right")

  # We set the font size of the axes in order to make the plots more readable
  ax[0].grid(axis="x", color="0.85")
  ax[0].grid(axis="y", color="0.85")

  ax[1].grid(axis="x", color="0.85")
  ax[1].grid(axis="y", color="0.85")

  for label in (ax[0].get_xticklabels() + ax[0].get_yticklabels() + ax[1].get_xticklabels() + ax[1].get_yticklabels()):
	  label.set_fontsize(14)

  # Show the plots
  plt.show()
```

## Building the model

One model with kernel regularization and one without it

```python
##function input, hidden layer e output e kernel regularization

def getModelRegularized(nhl,n_columns):
  model = Sequential()
  # input layer
  model.add(Dense(n_columns, activation="sigmoid", input_dim =n_columns))
  # hidden layers
  model.add(Dense(nhl, activation="sigmoid",kernel_initializer="random_normal",bias_initializer="random_normal",kernel_regularizer=keras.regularizers.l2(0.01)))
  #model.add(Dense(128, activation='relu'))
  # output layer
  model.add(Dense(1, activation = "sigmoid",kernel_initializer="random_normal",bias_initializer="random_normal",kernel_regularizer=keras.regularizers.l2(0.01)))
  
  return model
```

```python
## without kernel regularzation
def getModelSimple(nhl,n_columns):
  model = Sequential()
  # input layer
  model.add(Dense(n_columns, activation="sigmoid", input_dim =n_columns))
  # hidden layers
  model.add(Dense(nhl, activation="sigmoid",kernel_initializer="random_normal",bias_initializer="random_normal"))
  #model.add(Dense(128, activation='relu'))
  # output layer
  model.add(Dense(1, activation = "sigmoid",kernel_initializer="random_normal",bias_initializer="random_normal"))
  
  return model
```

We will use the binary_crossentropy as loss function and the metrics will be accuracy
